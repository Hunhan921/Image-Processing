## Task 1

Stack together three 200x175 rectanges of a different colours (firstly, horizontally, then  vertically). For example, blue, green and red. Save it to 'horizontal.jpg' and 'vertical.jpg' files.

```python
#write your code here
# Create the individual rectangles
blue = np.zeros((175, 200, 3), dtype=np.uint8)
blue[:, :, 0] = 255  # Set blue channel to 255

green = np.zeros((175, 200, 3), dtype=np.uint8)
green[:, :, 1] = 255  # Set green channel to 255

red = np.zeros((175, 200, 3), dtype=np.uint8)
red[:, :, 2] = 255  # Set red channel to 255

# Stack horizontally
horizontal = np.hstack((blue, green, red))
# Stack vertically
vertical = np.vstack((blue, green, red))
# Save the images
cv2.imwrite('horizontal.jpg', horizontal)
cv2.imwrite('vertical.jpg', vertical)
```

## Task 3

Write a factorial function (e.g., f(x) = x!, where x! = 1 * 2 * 3 * ... (x-1) * x).  

```python
#write your code here
def factorial(x):
    if x == 0:
        return 1
    else:
        result = 1
        for i in range(1, x + 1):
            result *= i
        return result
```

```python
factorial(4)
```
